```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import os

# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline
```

```python
birds = np.fromfile("bird_50_rdx.txt", sep='\n')
birds = birds.reshape(50,64)
print birds.shape
avg_birds = np.average(birds, axis=0)
max_birds = np.max(birds,axis=0)
```

```python
b1k = np.fromfile("birds_1k.txt", sep='\n')
b1k = b1k.reshape(1000, 64)
avg1k = np.average(b1k, axis=0)
print avg1k.shape
```

```python
for j in range(50):
    plt.plot([i for i in range(64)], birds[j])
plt.show()
```

```python
parks = np.fromfile("park_50_rdx.txt", sep="\n")
parks = parks.reshape(50,64)
avg_parks = np.average(parks, axis=0)
max_parks = np.max(parks, axis=0)

for j in range(50):
    plt.plot([i for i in range(64)], parks[j])
plt.show()
```

```python
p1k = np.fromfile("birds_1k.txt", sep='\n')
p1k = p1k.reshape(1000, 64)
avg1kp = np.average(p1k, axis=0)
print avg1kp.shape
```

```python
plt.plot([i for i in range(64)], avg_birds)
plt.plot([i for i in range(64)], avg_parks)
plt.show()

plt.plot([i for i in range(64)], avg1k)
plt.plot([i for i in range(64)], avg1kp)
plt.show()

plt.plot([i for i in range(64)], max_birds)
plt.plot([i for i in range(64)], max_parks)
plt.show()
```
