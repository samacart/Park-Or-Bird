```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import os

# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline
```

```python
birds = np.fromfile("..\\Data\\bird_50_rdx.txt", sep='\n')
birds = birds.reshape(50,64)
print birds.shape
avg_birds = np.average(birds, axis=0)
max_birds = np.max(birds,axis=0)
```

```python
b1k = np.fromfile("..\\Data\\birds_1k.txt", sep='\n')
b1k = b1k.reshape(1000, 64)
avg1k = np.average(b1k, axis=0)
print avg1k.shape
```

```python
for j in range(50):
    plt.plot([i for i in range(64)], birds[j])
plt.show()
```

```python
parks = np.fromfile("..\\Data\\park_50_rdx.txt", sep="\n")
parks = parks.reshape(50,64)
avg_parks = np.average(parks, axis=0)
max_parks = np.max(parks, axis=0)

for j in range(50):
    plt.plot([i for i in range(64)], parks[j])
plt.show()
```

```python
fig = plt.figure(figsize=(5,10))
fig.suptitle("Sample of 50 Parks and Birds with RGB ratio reduction", fontsize=14, fontweight='bold')
plt.subplot(2,1,1)
plt.title("50 Birds")
plt.xlabel("Block Position")
plt.ylabel("RGB Ratio")
for j in range(50):
    plt.plot([i for i in range(64)], birds[j])
plt.subplot(2,1,2)
plt.title("50 Parks")
plt.xlabel("Block Position")
plt.ylabel("RGB Ratio")
for j in range(50):
    plt.plot([i for i in range(64)], parks[j])
fig.show()
```

```python
p1k = np.fromfile("..\\Data\\parks_1k.txt", sep='\n')
p1k = p1k.reshape(1000, 64)
avg1kp = np.average(p1k, axis=0)
max_parks = np.max(parks, axis=0)
print avg1kp.shape
```

```python
fig = plt.figure(figsize=(5,10))
fig.suptitle("Means from two samples (50, 1000)", fontsize=14, fontweight='bold')
plt.subplot(2,1,1)
plt.title("Means from sample of 50")
plt.plot([i for i in range(64)], avg_birds, label="Mean 50 Birds")
plt.plot([i for i in range(64)], avg_parks, label="Mean 50 Parks")
plt.xlabel("Block Position")
plt.ylabel("RGB Ratio")
legend = plt.legend(loc='upper center', shadow=False)

plt.subplot(2,1,2)
plt.title("Means from sample of 1000")
plt.plot([i for i in range(64)], avg1k, label="Mean 1000 Birds")
plt.plot([i for i in range(64)], avg1kp, label="Mean 1000 Parks")
plt.xlabel("Block Position")
plt.ylabel("RGB Ratio")
legend = plt.legend(loc='upper center', shadow=False)
fig.show()

```

```python
fig = plt.figure(figsize=(5,5))
fig.suptitle("Max from 1000 Samples", fontsize=14, fontweight='bold')
plt.subplot(1,1,1)
plt.plot([i for i in range(64)], max_birds, label="Max 1000 Birds")
plt.plot([i for i in range(64)], max_parks, label="Max 1000 Parks")
plt.xlabel("Block Position")
plt.ylabel("RGB Ratio")
plt.legend(loc='upper left', shadow=False)
fig.show()
```
